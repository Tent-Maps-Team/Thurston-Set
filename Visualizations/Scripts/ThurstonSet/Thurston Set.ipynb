```python
import csv
import numpy as np 
import itertools 
import pandas as pd
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(25, 25))
plt.xlim(-1.5, 1.5)
plt.ylim(-1.5, 1.5)

real = []
imag = []
with open("00_Data_23.txt") as Data:
    readCSV = csv.reader(Data, delimiter=',')
    next(readCSV)

    for row in readCSV:

        it_raw = row[2].strip('[').strip(']').strip(' ').split(',')
        it = [int(i) for i in it_raw]
        n = int(row[0])
        coe = [0 for i in range(n + 1)] 
        coe[0] = 1 
        coe[n] = -1 
        for j in it:
            sumit = it.index(j) 
            coe[j + 1] = coe[j + 1] + (-2) * (sumit%2 * (-2) + 1) 
        roots = np.roots(coe)
        for j in roots:
            real.append(j.real)
            imag.append(j.imag)
plt.scatter(real,imag,s=.3,marker='.',color = '#ff00ff',alpha = 0.5)

real = []
imag = []
with open("00_Data_22.txt") as Data:
    readCSV = csv.reader(Data, delimiter=',')
    next(readCSV)

    for row in readCSV:

        it_raw = row[2].strip('[').strip(']').strip(' ').split(',')
        it = [int(i) for i in it_raw]
        n = int(row[0])
        coe = [0 for i in range(n + 1)] 
        coe[0] = 1 
        coe[n] = -1 
        for j in it:
            sumit = it.index(j) 
            coe[j + 1] = coe[j + 1] + (-2) * (sumit%2 * (-2) + 1) 
        roots = np.roots(coe)
        for j in roots:
            real.append(j.real)
            imag.append(j.imag)
plt.scatter(real,imag,s=.3,marker='.',color = '#0000ff',alpha = 0.5)

real = []
imag = []
with open("00_Data_21.txt") as Data:
    readCSV = csv.reader(Data, delimiter=',')
    next(readCSV)

    for row in readCSV:

        it_raw = row[2].strip('[').strip(']').strip(' ').split(',')
        it = [int(i) for i in it_raw]
        n = int(row[0])
        coe = [0 for i in range(n + 1)] 
        coe[0] = 1 
        coe[n] = -1 
        for j in it:
            sumit = it.index(j) 
            coe[j + 1] = coe[j + 1] + (-2) * (sumit%2 * (-2) + 1) 
        roots = np.roots(coe)
        for j in roots:
            real.append(j.real)
            imag.append(j.imag)
plt.scatter(real,imag,s=.3,marker='.',color = '#00ffff',alpha = 0.5)

real = []
imag = []
with open("00_Data_20.txt") as Data:
    readCSV = csv.reader(Data, delimiter=',',alpha = 0.5)
    next(readCSV)

    for row in readCSV:

        it_raw = row[2].strip('[').strip(']').strip(' ').split(',')
        it = [int(i) for i in it_raw]
        n = int(row[0])
        coe = [0 for i in range(n + 1)] 
        coe[0] = 1 
        coe[n] = -1 
        for j in it:
            sumit = it.index(j) 
            coe[j + 1] = coe[j + 1] + (-2) * (sumit%2 * (-2) + 1) 
        roots = np.roots(coe)
        for j in roots:
            real.append(j.real)
            imag.append(j.imag)
plt.scatter(real,imag,s=.3,marker='.',color = '#00ff00',alpha = 0.5)

real = []
imag = []
with open("00_Data_19.txt") as Data:
    readCSV = csv.reader(Data, delimiter=',',alpha = 0.5)
    next(readCSV)

    for row in readCSV:

        it_raw = row[2].strip('[').strip(']').strip(' ').split(',')
        it = [int(i) for i in it_raw]
        n = int(row[0])
        coe = [0 for i in range(n + 1)] 
        coe[0] = 1 
        coe[n] = -1 
        for j in it:
            sumit = it.index(j) 
            coe[j + 1] = coe[j + 1] + (-2) * (sumit%2 * (-2) + 1) 
        roots = np.roots(coe)
        for j in roots:
            real.append(j.real)
            imag.append(j.imag)
plt.scatter(real,imag,s=.3,marker='.',color = '#ffff00',alpha = 0.5)

real = []
imag = []
with open("00_Data_18.txt") as Data:
    readCSV = csv.reader(Data, delimiter=',')
    next(readCSV)

    for row in readCSV:

        it_raw = row[2].strip('[').strip(']').strip(' ').split(',')
        it = [int(i) for i in it_raw]
        n = int(row[0])
        coe = [0 for i in range(n + 1)] 
        coe[0] = 1 
        coe[n] = -1 
        for j in it:
            sumit = it.index(j) 
            coe[j + 1] = coe[j + 1] + (-2) * (sumit%2 * (-2) + 1) 
        roots = np.roots(coe)
        for j in roots:
            real.append(j.real)
            imag.append(j.imag)
plt.scatter(real,imag,s=.3,marker='.', color = '#ff9900',alpha = 0.5)


real = []
imag = []
with open("00_Data_17_.txt") as Data:
    readCSV = csv.reader(Data, delimiter=',')
    next(readCSV)

    for row in readCSV:

        it_raw = row[2].strip('[').strip(']').strip(' ').split(',')
        it = [int(i) for i in it_raw]
        n = int(row[0])
        coe = [0 for i in range(n + 1)] 
        coe[0] = 1 
        coe[n] = -1 
        for j in it:
            sumit = it.index(j) 
            coe[j + 1] = coe[j + 1] + (-2) * (sumit%2 * (-2) + 1) 
        roots = np.roots(coe)
        for j in roots:
            real.append(j.real)
            imag.append(j.imag)
plt.scatter(real,imag,s=.3,marker='.',color = '#ff0000',alpha = 0.5)





```
